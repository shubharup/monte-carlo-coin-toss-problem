```python
import pandas as pd
```

```python
import numpy as np

def uniform(n, m):
  return np.random.randint(1, n+1, size = m)
```

```python
def all_tosses_gen(iterations):
  return(uniform(2, 3*iterations))

# heads is 1, tails is 2
```

```python
def toss_score(toss):
  first_heads = None
  toss_score = -1

  for j in toss.keys():
    if toss[j] == 1:
      first_heads = j
      break

  if first_heads != None:
    toss_score = first_heads

  return(toss_score)
```

```python
def count_heads(toss):
  count = 0

  for j in toss.keys():
    if toss[j] == 1:
      count += 1

  return(count)
```

```python
def monte_carlo_estimate(rounds, all_tosses):
    no = 0

    for t in range(rounds):

        toss_triplet = {1: all_tosses[3*t], 2: all_tosses[3*t + 1], 3: all_tosses[3*t + 2]}

        if (count_heads(toss_triplet) < 3) and (toss_score(toss_triplet) <= 1):
            no = no + 1

    mc_probability = no / rounds
    return(mc_probability)
```

```python
rounds = 10000
all_tosses = all_tosses_gen(rounds)
```

```python
columns = ['N_points', 'estimate', 'error']
df_to_pickle = pd.DataFrame(columns=columns)
```

```python
for t in range(1,rounds):  # Adjust the range as needed

    estimated = monte_carlo_estimate(t, all_tosses)
    actual = 0.5

    row_data = {'N_points': t, 'estimate': estimated, 'error': (abs(estimated - actual)/actual)*100}

    df_to_pickle = df_to_pickle.append(row_data, ignore_index=True)
```

```python
df_to_pickle
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the pickled DataFrame from the file
df = df_to_pickle

# Calculate the trend line using numpy's polyfit
coefficients = np.polyfit(df['N_points'], df['error'], 1)
trend_line = np.polyval(coefficients, df['N_points'])

# Create an interactive scatter plot with trend line using Plotly Express
fig = px.scatter(df, x='N_points', y='error', title='Error vs N_points with Trend Line',
                 labels={'N_points': 'N_points', 'error': 'Error'},
                 hover_name='N_points', hover_data={'error': True, 'N_points': True})

# Add the trend line as a separate trace, ensuring it's on top
fig.add_trace(go.Scattergl(x=df['N_points'], y=trend_line, mode='lines', name='Trend Line', line=dict(color='red')))
fig.show()
```
